```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
xls = pd.ExcelFile('input_total/2024 12 03 Viajes almacen bcn solo mercancia pasada por volumétrica con clonados, ADR y Aduanas.xlsx')
viajes_df = pd.read_excel(xls, 'Viajes')
partidas_df = pd.read_excel(xls, 'Partidas')
```

```python
volumetrica_ultima_semana = partidas_df[(partidas_df['FechaEntradaAlmacen'] > '2024-12-09')]
volumetrica_counts = volumetrica_ultima_semana['TratadaVOlumetrica'].value_counts().rename({0: 'NO', 1: 'SI'})
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,5))

ax.pie(
    volumetrica_counts,
    labels = volumetrica_counts.index,
    startangle = 90,
    autopct = '%1.1f%%',
    pctdistance = 0.85
)

center_circle_2 = plt.Circle((0, 0), 0.70, fc='white')  # Set size and color
ax.add_artist(center_circle_2)

ax.set_title("Distribución tratado volumétrica última semana")
# Display the chart
plt.tight_layout()  # Ensure everything fits in the figure
plt.show()
```

```python
clonados_counts = partidas_df['DatosClonados'].value_counts()

fig, ax = plt.subplots(1,1,figsize=(5,5))

ax.pie(
    clonados_counts,
    labels = clonados_counts.index,
    startangle = 90,
    autopct = '%1.1f%%',
    pctdistance = 0.85
)

center_circle_2 = plt.Circle((0, 0), 0.70, fc='white')  # Set size and color
ax.add_artist(center_circle_2)

ax.set_title("Distribución clonados última semana")
# Display the chart
plt.tight_layout()  # Ensure everything fits in the figure
plt.show()
```

```python
remontables_ultima_semana = partidas_df[(partidas_df['TratadaVOlumetrica'] == 1) & (partidas_df['FechaEntradaAlmacen'] > '2024-12-09')]
remontable_counts = remontables_ultima_semana['Remontable'].value_counts()

fig, ax = plt.subplots(1,1,figsize=(5,5))

ax.pie(
    remontable_counts,
    labels = remontable_counts.index,
    startangle = 90,
    autopct = '%1.1f%%',
    pctdistance = 0.85
)

center_circle_2 = plt.Circle((0, 0), 0.70, fc='white')  # Set size and color
ax.add_artist(center_circle_2)

ax.set_title("Distribución remontables última semana")
# Display the chart
plt.tight_layout()  # Ensure everything fits in the figure
plt.show()
```
